# ACCIDENTES DE TRANSITO

Importamos los dataset de los lesioandos y fallecidos durante los años 2015-2019

```python
import pandas as pd
```

```python
#Columnas de los datos a utilizar
ColumnsFallecios=['año_ocu', 'día_ocu', 'hora_ocu', 'g_hora_5', 'mes_ocu', 'día_sem_ocu', 'depto_ocu', 'modelo_veh', 'tipo_eve', 'tipo_veh', 'fall_les', 'color_veh']
```

```python
fallecidos_2019 = pd.read_excel('Fallecidos-2019.xlsx', keep_default_na=False, usecols=ColumnsFallecios, index_col=0)
fallecidos_2019.head()
```

```python
fallecidos_2018 = pd.read_excel('Fallecidos-2018.xlsx', keep_default_na=False, usecols=ColumnsFallecios, index_col=0)
fallecidos_2018.head()
```

```python
fallecidos_2017 = pd.read_excel('Fallecidos-2017.xlsx', keep_default_na=False, usecols=ColumnsFallecios, index_col=0)
fallecidos_2017.head()
```

```python
fallecidos_2016 = pd.read_excel('Fallecidos-2016.xlsx', keep_default_na=False, usecols=ColumnsFallecios, index_col=0)
fallecidos_2016.head()
```

```python
fallecidos_2015 = pd.read_excel('Fallecidos-2015.xlsx', keep_default_na=False, usecols=ColumnsFallecios, index_col=0)
fallecidos_2015.head()
```

```python
DataFallecidos = [fallecidos_2019, fallecidos_2018, fallecidos_2017, fallecidos_2016, fallecidos_2015]
completeFallecidos = pd.concat(DataFallecidos, sort='True')
```

Generamos un csv, con los datos completos de los Fallecidos y Lesionados de los años 2015-2019

```python
completeFallecidos.to_csv('FallecidosCompletos.csv', sep=',')
```

```python
data = pd.read_csv('FallecidosCompletos.csv', na_values=['NaN'])
data.head()
```
